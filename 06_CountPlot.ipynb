# Countplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x = "sex", data = var)
plt.show()
```

```python
sns.barplot(x = "sex", y= "size", data = var)
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker")
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker", palette = "bwr")
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker", palette='dark:r', saturation = 0.5)
plt.show()
```
