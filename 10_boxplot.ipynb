# BoxPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var)
plt.show()
```

```python
sns.set(style = "whitegrid") # white, dark , whitegrid , darkgrid , ticks
sns.boxplot(x = "day" , y = "total_bill" , data = var)
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "time")
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "time", palette = "dark:g")
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "time", order = ["Fri", "Sun", "Sat", "Thur"])
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "sex", showmeans = True)
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "sex", showmeans = True , meanprops = {"marker": "+", 
           "markerfacecolor" :"g" , "markeredgecolor": "r"})
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "sex", linewidth = 5)
plt.show()
```

```python
sns.boxplot(x = "day" , y = "total_bill" , data = var, hue = "sex", linewidth = 5, palette = "plasma")
plt.show()
```

```python
sns.boxplot(y = "day" , x = "total_bill" , data = var, hue = "sex")
plt.show()
```

```python
sns.boxplot(data = var, orient = "h")
plt.show()
```

```python
sns.set(style = "whitegrid")
sns.boxplot(y = var["total_bill"])
plt.show()
```
