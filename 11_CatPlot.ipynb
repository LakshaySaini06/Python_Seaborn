# Cat Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips").head(50)
var
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "strip", hue = "sex") # strip , box , violin , boxen , point , bar , count, swarm
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "violin", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "box", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "boxen", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "point", hue = "sex") # earlier called factor plot
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "bar", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , data = var, kind = "count", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "swarm", hue = "sex")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "swarm", hue = "sex", palette = "BuPu")
plt.show()
```

```python
sns.catplot(x="size" , y ="tip" , data = var, kind = "swarm", hue = "sex", palette = "PuBu", height = 10)
plt.show()
```
