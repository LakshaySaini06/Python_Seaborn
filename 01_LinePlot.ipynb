# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7,8,9]
var_1 = [32,43,24,46,54,32,15,31,98]

x_1 = pd.DataFrame({"var": var , "var_1": var_1})
sns.lineplot(x = "var", y = "var_1", data = x_1)
```

```python
#matplotlib
plt.plot(var, var_1)
plt.show()
```

```python
sns.lineplot(x = var, y = var_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_1, hue = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_1, hue = "sex", style = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_1, hue = "sex", style = "sex", palette = 'Accent')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_1, hue = "sex", style = "sex", markers = ["o", ">"])
plt.show()
```

```python
y_2 = sns.load_dataset("penguins").head(20)
y_2
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_2, hue = "sex", style = "sex", markers = ["o", ">"], dashes = False)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_2, hue = "sex",legend  = "auto")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y = "bill_depth_mm" , data = y_2, hue = "sex",legend  = "full")
plt.grid()
plt.title("Lakshay")
plt.show()
```
