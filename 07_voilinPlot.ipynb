# Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "time")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, linewidth = 2, hue = "time")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "time" , palette = "PuBu")
plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill", data = var, order = ["Dinner", "Lunch"])
plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill", data = var, order = ["Dinner", "Lunch"], saturation = 1)
plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill", data = var, order = ["Dinner", "Lunch"], color = "r")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var,hue = "sex", split = True , density_norm = "width")
plt.show()
```

```python
sns.violinplot(y = "day", x = "total_bill", data = var,hue = "sex", split = True)
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var,hue = "sex", inner = None)# quartile, box , point ,stick, None
plt.show()
```

```python
sns.violinplot(y = var["total_bill"]) # can't make single violin plot for categorial or string data , it needs numerical values
plt.show()
```
