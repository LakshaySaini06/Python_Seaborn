# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x = var.island, y = var.bill_depth_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex")
plt.show()
```

```python
order_1 = ["Dream" ,"Biscoe","Torgersen" ]
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex", order = order_1, hue_order = ["Female", "Male"])
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex", errorbar = ("ci",90))
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex", errorbar = ("ci",90), n_boot = 100)
plt.show()
```

```python
sns.barplot(x = "bill_length_mm", y = "bill_depth_mm", data = var , hue = "sex",orient = "h")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",palette = "dark:blue")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",palette = "dark:r", saturation = 10)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",palette = "dark:green", err_kws = {"color": "red","linewidth" : 12})
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",capsize = 2)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",dodge = False)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",dodge = True, alpha = 0.5)
plt.show()
```

```python
sns.set(style = "darkgrid")
sns.barplot(x = "island", y = "bill_depth_mm", data = var , hue = "sex",dodge = True, alpha = 0.5)
plt.show()
```
