# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket_r", linewidth = 1, edgecolor = "g")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket_r",jitter = 5)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket_r",size = 20)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket_r",marker = "*")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket_r",alpha = 0.5)
plt.show()
```

```python
sns.stripplot(y = var["total_bill"])
plt.show()
```
