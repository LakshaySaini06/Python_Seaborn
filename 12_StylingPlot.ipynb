# Styling Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set_style("darkgrid") # dark , white , darkgrid , whitegrid , ticks
sns.boxplot(y =var["total_bill"])
plt.show()
```

```python
sns.boxplot(y =var["total_bill"])
sns.despine()
plt.show()
```

```python
sns.boxplot(y =var["total_bill"])
plt.show()
```

```python
plt.figure(figsize = (12, 5))
sns.boxplot(y =var["total_bill"])
plt.show()
```

```python
sns.set_context("poster", font_scale = 0.5)
sns.barplot(x = "day", y  ="total_bill" , data = var)
plt.show()
```

```python
sns.set_context("poster", font_scale = 0.5)
sns.barplot(x = "day", y  ="total_bill" , hue = "sex" , data = var, palette = "cool")
plt.show()
```
