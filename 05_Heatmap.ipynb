# HeatMap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = np.linspace(1,10,20).reshape(5,4)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
data
```

```python
data = sns.load_dataset("anagrams").head(10)
x = data.drop(columns = ["attnr"],axis = 1) 
sns.heatmap(x)
plt.show()
```

```python
sns.heatmap(x, vmin = 0, vmax = 12)
plt.show()
```

```python
sns.heatmap(x, vmin = 0, vmax = 12, cmap = "PuOr")
plt.show()
```

```python
sns.heatmap(x, vmin = 0, vmax = 12, cmap = "PuOr", annot = True)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
ar = [["L0","L1","L2","L3","L4"],
      ["P0","P1","P2","P3","P4"]] # array that is used as annot must have same shape as the data
sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = ar , fmt = 's')
plt.show()
```

```python
y = {"fontsize" : 20, "color" : "r"}
sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = ar , fmt = 's', annot_kws = y)
plt.show()
```

```python
sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = ar , fmt = 's', annot_kws = y, linewidth = 10, linecolor = "g")
plt.show()
```

```python
sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = ar , fmt = 's', annot_kws = y, cbar = False)
plt.show()
```

```python
sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = ar , fmt = 's', annot_kws = y, xticklabels = False, yticklabels = False)
plt.show()
```

```python
v = sns.heatmap(var1, vmin = 0, vmax = 12, cmap = "PuOr", annot = True)
v.set(xlabel = "Lakshay", ylabel = "Python")
sns.set(font_scale = 10)
plt.show()
```
