# Pair Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, hue = "sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["tip","total_bill"] ,hue = "sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["tip","total_bill"], hue = "sex", hue_order = ["Female", "Male"])
plt.show()
```

```python
sns.pairplot(var, vars = ["tip","total_bill"], hue = "sex", hue_order = ["Female", "Male"], palette = "BuGn")
plt.show()
```

```python
sns.pairplot(var, x_vars = ["tip","total_bill"], hue = "sex", hue_order = ["Female", "Male"])
plt.show()
```

```python
sns.pairplot(var, x_vars = ["tip","total_bill"], hue = "sex", hue_order = ["Female", "Male"], kind = "scatter")
plt.show()
```

```python
sns.pairplot(var,hue = "sex", markers = ["*","<"])
plt.show()
```
