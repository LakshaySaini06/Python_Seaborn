# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var =sns.load_dataset("penguins")
var
```

```python
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200, 210, 220 , 230])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200, 210, 220 , 230], kde = True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200, 210, 220 , 230], rug = True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200, 210, 220 , 230], kde = True, rug = True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200, 210, 220 , 230], color = "g")
plt.show()
```
