# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var,hue = "sex", style = "sex",size = "sex", sizes = (1000,40))
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var,hue = "sex", style = "sex",size = "sex", sizes = (1000,40), palette = "gist_ncar",
               alpha = 0.3)
plt.show()
```

```python
m = {"Male": "+", "Female": "x"}
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var,hue = "sex", style = "sex",size = "sex", markers = m)
plt.show()
```
